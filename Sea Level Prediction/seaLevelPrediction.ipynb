<h2> You will analyze a dataset of the global average sea level change since 1880. You will use the data to predict the sea level change through year 2050.

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
data = pd.read_csv('epa-sea-level.csv')
data.head(2)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Year'], data['CSIRO Adjusted Sea Level'], color='blue', label='Data')
```

```python
# Create first line of best fit (from 1880 to 2050)
slope, intercept, _, _, _ = linregress(data['Year'], data['CSIRO Adjusted Sea Level'])
years_extended = pd.Series(range(1880, 2051))
line1 = slope * years_extended + intercept
plt.plot(years_extended, line1, 'r', label='Best fit line (1880-2050)')
```

```python
 # Create second line of best fit (from 2000 to 2050)
recent_data = data[data['Year'] >= 2000]
slope_recent, intercept_recent, _, _, _ = linregress(recent_data['Year'], recent_data['CSIRO Adjusted Sea Level'])
years_recent = pd.Series(range(2000, 2051))
line2 = slope_recent * years_recent + intercept_recent
plt.plot(years_recent, line2, 'green', label='Best fit line (2000-2050)')
```
