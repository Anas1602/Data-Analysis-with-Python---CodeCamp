<h1> Read data from SQL database

```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect('chinook.db')
```

```python
cur = conn.cursor()
```

```python
cur.execute('SELECT * FROM employees LIMIT 5;')
```

```python
results = cur.fetchall()
results
```

```python
df = pd.DataFrame(results)
df.head()
```

<h3> Using pandas read_sql method

```python
df = pd.read_sql('SELECT * FROM employees;', conn)
df
```

```python
df = pd.read_sql('SELECT * FROM employees;', conn,
                 index_col='EmployeeId',
                 parse_dates=['BirthDate', 'HireDate'])
df.head()
```

```python
df.info()
```

```python
df['ReportsTo'].isna().sum()
```

```python
df['ReportsTo'] > 1.75
```

<h3> Using pandas read_sql_query method

```python
conn = sqlite3.connect('chinook.db')
df = pd.read_sql_query('SELECT * FROM employees LIMIT 5;', conn)
df.head()
```

<h3> Using read_sql_table method

```python
from sqlalchemy import create_engine
engine = create_engine('sqlite:///chinook.db')

connection = engine.connect()
df = pd.read_sql_table('employees', con=connection)
df.head()
```

<h3> Create tables from DataFrame objects

```python
df.to_sql('employees2', conn)
```

```python
pd.read_sql_query('SELECT * FROM employees2;', conn).head()
```

```python
pd.DataFrame().to_sql('employees2',
                      conn,
                      if_exists='replace')
```

```python
pd.read_sql_query('SELECT * FROM employees2;', conn).head()
```

```python
df.to_sql('employees2',
          conn,
          if_exists='replace')
```

```python
pd.read_sql_query('SELECT * FROM employees2;', conn).head()
```
